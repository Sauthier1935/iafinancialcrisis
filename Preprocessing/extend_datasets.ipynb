# (3) Data Normalization

### Separate instances from each country to fill in, together with the country identification.

### seen in the previous item that rconpc, gdp, iy, ca, imports, exports, narrowm, monet, stir, ltrate, stocks, debtgdp, revenue, expenditure, xrusd, 2331, tloans, tmort, thh, tbus, hpnom

#### all_columns = columns that will not be used in the empty cells filling process
#### df_aux = dataframe with the information that will be normalized


### Separar instâcias de cada país para realizar preenchimento, junto a identificação do país.

### oservado no item anterior que rconpc, gdp, iy, ca, imports, exports, narrowm, monet, stir, ltrate, stocks, debtgdp, revenue, expenditure, xrusd, 2331, tloans, tmort, thh, tbus, hpnom

#### all_columns = colunas que não serão utilizadas no processo de preenchimento das células vazia
#### df_aux = dataframe com as informações que serão normalizadas

```
# Retirar tipo objetos e instancias imutaveis e que possuem preenchimento completo.
all_columns = ['year', 'iso','country', 'ifs', 'pop','rgdpmad','rgdppc','cpi','crisisJST']

# year, country, ifs, pop,rgdpmad,rgdppc,cpi,CrisisJST and crisisJST
df_aux = df_crisis.drop(['year', 'country', 'ifs', 'pop','rgdpmad','rgdppc','cpi','crisisJST'],axis=1)
```

### Metodo utilizado com a normalizalção e a estratégia

#### imp_mean = simple imputer normalizing function
#### imp_median = simple imputer normalizing function
#### imp_most_f = simple imputer normalizing function

#### imp_mean = função normalizadora simple imputer 
#### imp_median = função normalizadora simple imputer 
#### imp_most_f = função normalizadora simple imputer 

```
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean') #normalizador é a media do atributo
imp_median = SimpleImputer(missing_values=np.nan, strategy='median') #normalizador é a mediana do atributo
imp_most_f = SimpleImputer(missing_values=np.nan, strategy='most_frequent') #normalizador é a celular mais frequente do atributo
```

### função que realiza normalização

#### iso_country = nome dos 17 países abreviados

```
iso_country = {'iso': ['AUS','BEL','CAN','CHE','DEU','DNK','ESP','FIN','FRA','GBR','ITA','JPN','NLD','NOR','PRT','SWE','USA'],}
iso_country = pd.DataFrame(iso_country)
```

```
def normalize_mean(df, iso):
  
  df =df.drop(['iso'], axis = 1)
  c_names = df.columns.tolist() #separa nome das colunas
  imp_mean.fit(df)
  SimpleImputer()
  df = pd.DataFrame(imp_mean.transform(df)) #preencher celular vazias
  df.columns = c_names
  df['iso'] = iso

  return df
```

```
aus = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][0]], iso_country['iso'][0])
bel = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][1]], iso_country['iso'][1])
can = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][2]], iso_country['iso'][2])
che = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][3]], iso_country['iso'][3])
deu = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][4]], iso_country['iso'][4])
dnk = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][5]], iso_country['iso'][5])
esp = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][6]], iso_country['iso'][6])
fin = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][7]], iso_country['iso'][7])
fra = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][8]], iso_country['iso'][8])
gbr = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][9]], iso_country['iso'][9])
ita = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][10]], iso_country['iso'][10])
jpn = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][11]], iso_country['iso'][11])
nld = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][12]], iso_country['iso'][12])
nor = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][13]], iso_country['iso'][13])
prt = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][14]], iso_country['iso'][14])
swe = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][15]], iso_country['iso'][15])
usa = normalize_mean(df_aux.loc[df_aux['iso'] == iso_country['iso'][16]], iso_country['iso'][16])
```

### concatenar todos os países em um único dataframe
#### frames = todos os dfs de cada pais
#### union = concatenação dos frames

```
frames = [aus,bel,can,che,deu,dnk,esp,fin,fra,gbr,jpn,nld,nor,prt,swe,usa]
union_mean = pd.concat(frames, ignore_index=True)
union_mean.reset_index(drop=True)
union_mean[all_columns] = df_crisis[all_columns] # recebe atributos não modificados
```

```
df = pd.DataFrame(union_mean)
df.to_excel(r'/content/drive/MyDrive/Mean.xlsx', index = False)
```
